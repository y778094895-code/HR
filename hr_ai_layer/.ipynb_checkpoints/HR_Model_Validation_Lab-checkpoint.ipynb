## 1.4 Target Engineering هندسة الهدف (Decline Detection)

```python
# ==========================================
# ترتيب البيانات زمنياً لكل موظف
# ==========================================

df_perf["record_date"] = pd.to_datetime(df_perf["record_date"])
df_perf = df_perf.sort_values(["employee_id", "record_date"])

# Lag features للأداء
df_perf["lag1"] = df_perf.groupby("employee_id")["overall_score"].shift(1)
df_perf["lag2"] = df_perf.groupby("employee_id")["overall_score"].shift(2)

df_perf["diff1"] = df_perf["overall_score"] - df_perf["lag1"]
df_perf["diff2"] = df_perf["lag1"] - df_perf["lag2"]

df_perf["employee_mean"] = df_perf.groupby("employee_id")["overall_score"].transform("mean")

# تعريف التراجع الواقعي
sustained_drop = (df_perf["diff1"] < -0.3) & (df_perf["diff2"] < -0.3)
relative_drop = (df_perf["overall_score"] < df_perf["employee_mean"] * 0.90)

df_perf["decline_flag"] = (sustained_drop | relative_drop).astype(int)

df_perf = df_perf.dropna().copy()

print("Decline Rate:", df_perf["decline_flag"].mean())
```

### ماذا فعلنا هنا؟

- رتبنا البيانات زمنيًا لكل موظف.
- أنشأنا Lag features.
- حسبنا فروقات الأداء.
- عرفنا التراجع بطريقتين:
  - انخفاض مستمر
  - انخفاض نسبي عن المتوسط الشخصي
- أنشأنا عمود decline_flag كهدف للنموذج.

# ✅ LAYER 2 — MODEL TRAINING ENGINE
# ✅ الطبقة الثانية — محرك تدريب النماذج

في هذه الطبقة سنقوم بـ:
- بناء ميزات زمنية (Lag / Trend / Volatility)
- تجهيز بيانات التدريب والاختبار بشكل صارم (Forward split)
- تدريب أكثر من خوارزمية ومقارنتها
- اختيار النموذج الأفضل رسميًا بناءً على AUC + F1

## 2.1 Feature Engineering هندسة الميزات المتقدمة

```python
# ==========================================
# إنشاء ميزات زمنية إضافية
# ==========================================

def build_features(df):
    df = df.copy()

    feature_cols = [
        "kpi_productivity_mean",
        "kpi_quality_mean",
        "burnout_index_mean",
        "motivation_index",
        "overtime_hours_mean",
        "workload",
    ]

    for col in feature_cols:
        df[f"{col}_lag1"] = df.groupby("employee_id")[col].shift(1)
        df[f"{col}_trend3"] = (
            df.groupby("employee_id")[col]
            .rolling(3)
            .mean()
            .reset_index(level=0, drop=True)
        )
        df[f"{col}_vol3"] = (
            df.groupby("employee_id")[col]
            .rolling(3)
            .std()
            .reset_index(level=0, drop=True)
        )

    df = df.fillna(0)

    blacklist = [
        "employee_id",
        "record_date",
        "decline_flag",
        "lag1",
        "lag2",
        "diff1",
        "diff2",
        "employee_mean"
    ]

    features = [c for c in df.columns if c not in blacklist]

    return df, features


train_df, features = build_features(train_df)
test_df, _ = build_features(test_df)

X_train = train_df[features]
y_train = train_df["decline_flag"]

X_test = test_df[features]
y_test = test_df["decline_flag"]

print("عدد الميزات:", len(features))
```

<div style="font-size:14px; line-height:1.7">

### ماذا فعلنا هنا؟ ✅  
- بنينا ميزات زمنية واقعية لكل موظف مثل:
  - Lag1: قيمة الشهر السابق  
  - Trend3: متوسط آخر 3 أشهر (اتجاه)  
  - Vol3: تذبذب آخر 3 أشهر (عدم استقرار)  
- نظّفنا البيانات من NaN و Inf.
- استخرجنا قائمة ميزات بدون أي تسريب (نستبعد الهدف والهوية والتاريخ و overall_score).

</div>

## 2.2 Model Candidates الخوارزميات المرشحة للتدريب

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier

# ✅ قائمة خوارزميات قوية وشائعة للمشاريع الإنتاجية
models = {
    "LogReg": LogisticRegression(max_iter=3000),
    "RF": RandomForestClassifier(
        n_estimators=400,
        min_samples_leaf=10,
        n_jobs=-1,
        random_state=42
    ),
    "HGB": HistGradientBoostingClassifier(
        max_depth=6,
        learning_rate=0.05,
        max_iter=600,
        random_state=42
    )
}

list(models.keys())
```

<div style="font-size:14px; line-height:1.7">

### ماذا فعلنا هنا؟ ✅  
- جهزنا 3 خوارزميات:
  - LogisticRegression (Baseline سريع ومفسر)
  - RandomForest (قوي مع التفاعلات غير الخطية)
  - HistGradientBoosting (قوي جدًا للبيانات الجدولية غالبًا)

</div>

## 2.3 Train + Evaluate (Time Holdout) تدريب وتقييم على اختبار زمني (بدون تسريب)

```python
from sklearn.metrics import roc_auc_score, f1_score, precision_recall_curve, classification_report, confusion_matrix

def best_threshold_by_f1(y_true, probs):
    # ✅ نبحث عن أفضل Threshold يحقق أعلى F1
    p, r, t = precision_recall_curve(y_true, probs)
    f1 = (2 * p * r) / (p + r + 1e-9)
    # t طوله أقل بواحد من p,r
    idx = np.argmax(f1[:-1])
    return float(t[idx])

results = {}

for name, model in models.items():
    # ✅ تدريب
    model.fit(X_train, y_train)

    # ✅ احتمالات
    probs = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else model.predict(X_test)

    # ✅ Threshold أفضل
    thr = best_threshold_by_f1(y_test, probs)
    preds = (probs >= thr).astype(int)

    # ✅ مقاييس
    auc = roc_auc_score(y_test, probs)
    f1 = f1_score(y_test, preds)

    results[name] = {
        "AUC": float(auc),
        "F1": float(f1),
        "Threshold": float(thr),
        "CM": confusion_matrix(y_test, preds)
    }

    print(f"\n--- {name} ---")
    print("AUC:", auc)
    print("F1 :", f1)
    print("Threshold:", thr)
    print("Confusion Matrix:\n", results[name]["CM"])
    print("\nClassification Report:\n", classification_report(y_test, preds))
```

<div style="font-size:14px; line-height:1.7">

### ماذا فعلنا هنا؟ ✅  
- دربنا كل خوارزمية على Training فقط.
- اختبرنا على Test زمني (مستقبل) لضمان الواقعية.
- حسبنا:
  - ROC-AUC: قوة الترتيب/التمييز
  - F1: جودة التوقع الفعلي بعد اختيار Threshold
- اخترنا Threshold بشكل علمي عبر منحنى Precision-Recall لتحقيق أعلى F1.

</div>

## 2.4 Select Champion Model اختيار النموذج الفائز رسميًا

في هذه المرحلة سنقوم بـ:

## 4.1 Risk Segmentation تقسيم مستويات المخاطر

```python
# ==========================================
# إنشاء DataFrame للتقييم
# ==========================================

df_eval = pd.DataFrame({
    "probability": probs,
    "actual": y_test
})

# ==========================================
# تقسيم المخاطر إلى 3 شرائح
# ==========================================

df_eval["risk_band"] = pd.cut(
    df_eval["probability"],
    bins=[0, 0.20, 0.40, 1.0],
    labels=["Low", "Medium", "High"],
    include_lowest=True
)

df_eval["risk_band"].value_counts()
```

<div style="font-size:14px; line-height:1.7">

### ماذا فعلنا هنا؟

- حولنا الاحتمالات إلى شرائح مفهومة إداريًا.
- Low → مخاطر منخفضة
- Medium → تحتاج مراقبة
- High → تحتاج تدخل فوري

الآن النموذج أصبح مفهومًا لمدير الموارد البشرية.

</div>

## 4.2 Actual Decline Rate per Segment معدل التراجع الفعلي داخل كل شريحة

```python
segment_analysis = df_eval.groupby("risk_band", observed=True)["actual"].mean()
segment_analysis
```

<div style="font-size:14px; line-height:1.7">

### ماذا نقرأ هنا؟

إذا كانت:

- Medium → متوسط  
- Low → منخفض

</div>

## 4.3 Economic Simulation محاكاة الأثر المالي للتدخل

```python
# ==========================================
# افتراضات مالية
# ==========================================

COST_OF_DECLINE = 15000        # خسارة تقديرية عند تراجع أداء موظف
INTERVENTION_COST = 2000       # تكلفة تدخل إداري

def simulate_strategy(threshold):
    selected = df_eval[df_eval["probability"] >= threshold]

    targeted = len(selected)
    prevented = selected["actual"].sum()

    net_gain = (prevented * COST_OF_DECLINE) - (targeted * INTERVENTION_COST)

    return {
        "threshold": threshold,
        "targeted_employees": targeted,
        "actual_declines": int(prevented),
        "net_gain": int(net_gain)
    }

thresholds_to_test = [0.20, 0.30, 0.40, 0.50]

economic_results = pd.DataFrame(
    [simulate_strategy(t) for t in thresholds_to_test]
)

economic_results
```

<div style="font-size:14px; line-height:1.7">

### ماذا نفعل هنا؟

- نحاكي سيناريو تدخل إداري.
- نختبر عدة Threshold.
- نحسب:
  - عدد الموظفين المستهدفين.
  - عدد التراجعات الفعلية.
  - صافي الربح أو الخسارة.

هنا يتحول النموذج إلى أداة قرار حقيقية.

</div>

## 4.4 Optimal Business Threshold اختيار العتبة المثلى إداريًا

```python
best_strategy = economic_results.sort_values("net_gain", ascending=False).iloc[0]
best_strategy
```

<div style="font-size:14px; line-height:1.7">

## 5.3 Fairness & Bias Check فحص العدالة والانحياز

```python
df_eval["burnout_group"] = pd.qcut(
    df_eval["burnout_index"],
    4,
    labels=["Q1","Q2","Q3","Q4"]
)

fairness_analysis = df_eval.groupby(
    "burnout_group",
    observed=True
)["prediction"].mean()

fairness_analysis
```

#### ماذا نفعل هنا؟

نختار ميزة موجودة فعليًا في البيانات.

نقسمها إلى 4 شرائح (Quartiles)
ثم نرى متوسط التنبؤ داخل كل شريحة.

إذا كان النموذج متوازنًا → جيد.
إذا كان يعطي حكمًا متطرفًا لمجموعة معينة → هناك انحياز.
